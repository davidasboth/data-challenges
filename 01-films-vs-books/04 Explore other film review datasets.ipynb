`movies.csv` from https://www.kaggle.com/datasets/patkle/metacritic-scores-for-games-movies-tv-and-music

```python
import pandas as pd
```

```python
films = pd.read_csv("./data/raw/movies.csv")
print(films.shape)
films.head()
```

```python
duplicate_films = films["title"].value_counts().loc[lambda x: x > 1].index
films[films.title.isin(duplicate_films)].sort_values("title")
```
