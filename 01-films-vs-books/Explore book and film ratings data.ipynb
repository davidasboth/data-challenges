# Book ratings

From: https://github.com/zygmuntz/goodbooks-10k/tree/master (`books.csv`)

```python
import pandas as pd
```

```python
books = pd.read_csv("./data/raw/books.csv")
print(books.shape)
books.head()
```

```python
books.head(1).T
```

```python
book_ratings = books[
    [
        "goodreads_book_id",
        "authors",
        "original_publication_year",
        "original_title",
        "title",
        "average_rating",
        "small_image_url",
    ]
].copy()

book_ratings.head()
```

We need to normalise book ratings to /100. It's currently out of 5 so we can just multiply by 20.

```python
book_ratings["average_rating_normalised"] = book_ratings["average_rating"] * 20

book_ratings["authors"] = book_ratings["authors"].str.upper().str.strip()
book_ratings["original_title"] = book_ratings["original_title"].str.upper().str.strip()
book_ratings["title"] = book_ratings["title"].str.upper().str.strip()

book_ratings.rename(
    columns={
        "authors": "author",
        "original_title": "book_title",
        "average_rating_normalised": "avg_book_rating",
    }
)

book_ratings.head()
```

```python
book_ratings.to_parquet("./data/book_ratings.parquet", index=False)
```

# Films

From: https://www.kaggle.com/datasets/talha002/rottentomatoes-400k-review

```python
films = pd.read_csv("./data/raw/rottentomatoes-400k.csv")
print(films.shape)
films.head()
```

```python
sorted(films["Score"].unique())
```

```python
films[films["Score"] < 2]
```

On Rotten Tomatoes, ratings seem to be out of 5 or 4 or 100 or a letter system, but this data seems to be normalised to being /100

### Roll up films to film-level

We can only group by title, but what if two films have the same title?!?

How often does that happen?

If we look at the min and max years for each movie, do we have ones that have a range that suggests it's actually two movies?

```python
films["Date"] = pd.to_datetime(films["Date"], format="mixed", dayfirst=True)
films["year"] = films["Date"].dt.year

film_dates = films.groupby("Movie")["year"].agg(["min", "max"]).reset_index()
film_dates["range"] = film_dates["max"] - film_dates["min"]
film_dates.sort_values("range", ascending=False).head(20)
```

Looks like we do have films that have a long range between first and latest reviews but this could be:

- because people are still reviewing old films
- data errors (reviews from 1800)
- they are reviews about different films (we have no other way to tell)

Conclusion: we'll assume reviews of a movie title are always the same movie (for better or worse)

```python
film_ratings = films.groupby("Movie")["Score"].mean().reset_index()

film_ratings.columns = ["film_title", "avg_film_rating"]

film_ratings["film_title"] = film_ratings["film_title"].str.upper().str.strip()

film_ratings.head()
```

```python
film_ratings.to_parquet("./data/film_ratings.parquet", index=False)
```
